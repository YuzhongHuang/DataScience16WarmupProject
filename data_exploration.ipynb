## Yuzhong's Titanic Data Exploration 
    

```python
import numpy as np
import pandas as pd
import thinkstats2
import thinkplot

df = pd.read_csv("train.csv")
df
```

## Survival Rate among Pclass 1, 2, 3

```python
firsts = df[df.Pclass == 1]
seconds = df[df.Pclass == 2]
thirds = df[df.Pclass == 3]

print "first p class survival rate: ", firsts.Survived.mean()
print "second p class survival rate: ", seconds.Survived.mean()
print "third p class survival rate: ", thirds.Survived.mean()
```

From the means of firsts', seconds' and thirds' P class's survival rate. There might be some sort of correlation behind P class rank. We can see that the higher the rank, the higher the chance to survive Titanic.

## Survival Rate between man and woman

```python
women = df[df.Sex == "female"]
men = df[df.Sex == "male"]

print "female survival rate: ", women.Survived.mean()
print "male survival rate: ", men.Survived.mean()

df = df[]
```

Similarly, female seems to have a significant higher chance to survive in Titanic

## Survival rate of women among different P class ranks

```python
first_women = women[df.Pclass == 1]
second_women = women[df.Pclass == 2]
third_women = women[df.Pclass == 3]

print "first class female survival rate: ", first_women.Survived.mean()
print "second class female survival rate: ", second_women.Survived.mean()
print "third class female survival rate: ", third_women.Survived.mean()
```

It seems that first and second class women had a relatively high rate of survival, while third class only got half the chance

## Scatter Graph for Fare

```python
%matplotlib inline
thinkplot.Scatter(df.Fare, df.Survived)
thinkplot.Show(xlabel='Fare',
               ylabel='Survival Rate')
```

It seems that people with low fares seems to have a better chance of surviving. Since the blue color is very dense at low fare

## Binning the Grapn

```python
thinkplot.HexBin(df.Fare, df.Survived)
thinkplot.Show(xlabel='Fare',
               ylabel='Survival Rate',
               axis=[0,200,-0.2,1.2])
```

We can barely get information from the binned graph since data is not continuous(survival only have 0 and 1)
